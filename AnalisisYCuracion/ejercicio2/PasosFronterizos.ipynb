<h2>Ejercicio</h2>

* Graficar histograma de cantidad de pasos por provincia

<h3>1 Importando pandas, matplotlib, seaborn</h3>

```python
import pandas
import seaborn
import matplotlib.pyplot as plt
```

<h3>2 Cargando datos</h3>

```python
pasos_fronterizos = pandas.read_json("pasos.json")
pasos_fronterizos
```

<h3>3 Grafica</h3>

```python
# Configuración del grafico
plt.figure(figsize=(20,10))
plt.title('Histograma de cantidad por provincia')

# Optención de data frame con datos de cantidad de pasos por provincia
cantidad_de_pasos_por_provincia_data_frame = pasos_fronterizos.province.value_counts().to_frame()

# Grafico con seaborn
seaborn.distplot(cantidad_de_pasos_por_provincia_data_frame)
```
